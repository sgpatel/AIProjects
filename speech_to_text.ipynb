```
import sounddevice as sd
import numpy as np
import wavio
from pydub import AudioSegment

# Record parameters
duration = 5  # seconds
fs = 44100  # Sample rate

# Step 1: Record the audio
print("Recording...")
myrecording = sd.rec(int(duration * fs), samplerate=fs, channels=2)
sd.wait()  # Wait until recording is finished
print("Recording finished")

# Save the recording as WAV file
wavio.write("output.wav", myrecording, fs, sampwidth=2)

# Step 2: Convert the recorded WAV file to MP3
audio = AudioSegment.from_wav("output.wav")
audio.export("output.mp3", format="mp3")

print("Saved as output.mp3")
```

```
import whisper
from googletrans import Translator
from google.colab import files

# Step 1: Upload the audio file (WAV format)
uploaded = files.upload()

# Step 2: Load the Whisper model
model = whisper.load_model("medium")

# Step 3: Transcribe the audio in Hindi
file_name = list(uploaded.keys())[0]
result = model.transcribe(file_name, language='hi')  # Transcribing Hindi audio
hindi_text = result['text']

print("Transcribed Hindi Text:")
print(hindi_text)

# Step 4: Translate Hindi to English using Whisper's built-in translation capability
result_en = model.transcribe(file_name, task='translate')  # Translating directly to English
english_text = result_en['text']

print("\nTranslated English Text:")
print(english_text)

# Step 5: Translate Hindi to Marathi using Googletrans
translator = Translator()
marathi_text = translator.translate(hindi_text, src='hi', dest='mr').text

print("\nTranslated Marathi Text:")
print(marathi_text)
```
